# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекс Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Вам нужно научиться определять рыночную стоимость объектов недвижимости. Для этого проведите исследовательский анализ данных и установите параметры, влияющие на цену объектов. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность.

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта и других объектов — эти данные автоматически получены из геосервисов. Количество парков и водоёмов также заполняется без участия пользователя.
Описание данных
airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
balcony — число балконов
ceiling_height — высота потолков (м)
cityCenters_nearest — расстояние до центра города (м)
days_exposition — сколько дней было размещено объявление (от публикации до снятия)
first_day_exposition — дата публикации
floor — этаж
floors_total — всего этажей в доме
is_apartment — апартаменты (булев тип)
kitchen_area — площадь кухни в квадратных метрах (м²)
last_price — цена на момент снятия с публикации
living_area — жилая площадь в квадратных метрах (м²)
locality_name — название населённого пункта
open_plan — свободная планировка (булев тип)
parks_around3000 — число парков в радиусе 3 км
parks_nearest — расстояние до ближайшего парка (м)
ponds_around3000 — число водоёмов в радиусе 3 км
ponds_nearest — расстояние до ближайшего водоёма (м)
rooms — число комнат
studio — квартира-студия (булев тип)
total_area — общая площадь квартиры в квадратных метрах (м²)
total_images — число фотографий квартиры в объявлении

### Выполните предобработку данных

Создаем таблицу с количеством пропущенных значений.
Создаем таблицу с процентным соотношением пропущенных значений,объединяем таблицы в одну.

```python
missing_values = data.isnull().sum()
missing_percentages = (data.isnull().mean() * 100).round(2)
missing_info = pd.concat([missing_values, missing_percentages], axis=1)
missing_info.columns = ['Количество пропусков', 'Процент пропусков']
print(missing_info)
```

Пропуски в таблице могут быть вызваны ошибками при заполнении данных, 
отсутствием обязательных полей или недоступностью информации.
Процент пропусков важен для оценки качества данных и выбора метода их обработки.
Обращаю внимание, в данной таблице , есть столбцы с пропусками более 20%, замена или удаление данных с такими пропусками может привести к искажению данных    

Смотрим на уникальные значения в столбцах, 
чтобы обнаружить необычные данные и возможные ошибки.
Заполняем пропущенные значения в столбце balcony на 0, так как пропуски могут быть связаны с отсутствием балкона.
Так как большинство недвижимости не является апартаментами, хотя там и большой процент пропуска, заменим их на  False.
Заполним пустые значения медианными значениями living_area для каждой категории total_area,как наиболее зависимой, так же я округлил значение до целых чисел, что бы график был более наглядным    


```python
def count_value_occurrences(data, columns):
    for column in columns:
        print(f"Значения в столбце '{column}':")
        print(data[column].value_counts())
        print("\n")

columns_to_count = ['total_images', 'ceiling_height', 'floor', 'is_apartment', 'total_area', 
                    'rooms', 'airports_nearest', 'cityCenters_nearest', 'parks_around3000', 
                    'parks_nearest', 'ponds_around3000']
count_value_occurrences(data, columns_to_count)

data['balcony'].value_counts()
data['balcony'] = data['balcony'].fillna(0)

data['balcony'] = data['balcony'].astype('int')
data['is_apartment'] = data['is_apartment'].fillna(False)
data['locality_name'].fillna('нет имени',inplace=True)
data['living_area'] = data.groupby(pd.qcut(data['total_area'], 10))['living_area'].transform(lambda x: x.fillna(x.median()))
data['living_area'] = data['living_area'].round()
```

Для замены пропущенных значений в столбце 'days_exposition' была использована медиана. Выполнена замена пропусков медианами значениями сгруппированными по неделям, помогает учесть сезонные или недельные изменения в продолжительности размещения объявлений. 
Затем столбец 'days_exposition' был преобразован в целочисленный тип данных для удобства использования в дальнейшем анализе данных. Таким образом, мы минимизируем искажения и сохраняем распределение значений в столбце.

```python
data['first_day_exposition']=pd.DatetimeIndex(data['first_day_exposition'])
data['week'] = data['first_day_exposition'].dt.isocalendar().week
data['days_exposition'] = data['days_exposition'].fillna(data.groupby('week')['days_exposition'].transform('median'))
missing_values = data['days_exposition'].isnull().sum()
print("Количество пропущенных значений в столбце 'days_exposition':", missing_values)
data['days_exposition'] = data['days_exposition'].astype(int)
```

Алёна, если тебе не сложно, подскажи, где я могу найти дополнительный материал по обработке данных, особенно по замене пропусков.

Мы заменили пропуски в столбе kitchen_area медианам значением сгруппированных по столбу rooms. 
Что минимизирует искажения  которые могут возникнуть при  заполнении пропусков,если невозможно вычислить этот параметр ставим 0  

```python
data['kitchen_area'] = data.groupby('rooms')['kitchen_area'].transform(lambda x: x.fillna(x.median() if x.notnull().any() else 0))
data['kitchen_area'].isna().sum()
```

Пропуски в данных могут возникать по разным причинам, 
таким как отсутствие обязательной информации при размещении объявления,
утеря данных из-за технических проблем или человеческого фактора,
недоступность определенных параметров для заполнения, ошибки при выгрузке и обработке данных, #отсутствие предоставленной информации от продавца,
а также неопределенность или неизвестность некоторых параметров для объектов недвижимости. 
Но не все пропуски можно заменить, большой процент пропусков 
в столбце может привести к искажению результатов анализа данных

Устраняем неявные дубликаты в названиях населённых пунктов.


```python
locality_names_unique = data['locality_name'].str.lower().str.replace('ё', 'е').unique()
print(locality_names_unique)
data['locality_name'] = (
    data['locality_name']
    .str.replace('ё', 'е') 
    .str.replace('(городской поселок|поселок городского типа)', 'поселок', regex=True) )
```

```python
data['floors_total'] = data.groupby('locality_name')['floors_total'].transform(lambda x: x.fillna(x.median()))
```

```python
missing_values = data.isnull().sum()
missing_percentages = (data.isnull().mean() * 100).round(2)
missing_info = pd.concat([missing_values, missing_percentages], axis=1)
missing_info.columns = ['Количество пропусков', 'Процент пропусков']
print(missing_info)
```

```python
duplicates = data.duplicated()
print("Количество дубликатов:", duplicates.sum())
```

Было выполнена предобработка данных, пропущенные значения были заполнены или оставлены, где это было невозможно. Также были изменены типы данных в некоторых столбцах и устранены неявные дубликаты в столбце с названиями населенных пунктов, проверено отсутствуе дубликатов.

### Добавьте в таблицу новые столбцы

Шаг 3
Добавим в таблицу новые столбцы со следующими параметрами:
цена одного квадратного метра 
(нужно поделить стоимость объекта на его общую площадь,
а затем округлить до двух знаков после запятой);

```python
data['price_per_sqm'] = (data['last_price'] / data['total_area']).round(2)
```

день недели публикации объявления (0 — понедельник, 1 — вторник и так далее);
месяц публикации объявления;
год публикации объявления;

```python
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday
data['month_exposition'] = data['first_day_exposition'].dt.month
data['year_exposition'] = data['first_day_exposition'].dt.year
```

тип этажа квартиры (значения — «первый», «последний», «другой»);

```python
def floor_type(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
data['floor_type'] = data.apply(floor_type, axis=1)
floor_type_counts = data['floor_type'].value_counts()
print(floor_type_counts)
```

Переводим расстояние до центра города из метров в километры и округляем

```python
data['cityCenters_nearest_km'] = (data['cityCenters_nearest'] / 1000).round()
```

Исходя из графиков подсчетам часто встречающихся значений в диапазоне от 0 до 100
В столбе days_exposition  мы заменили аномальные значения 0, 7, 30, 45,60,90, медианами значениями сгруппированными по неделям.

```python
data.plot(y = 'days_exposition', kind='hist', bins = 100, figsize=(5, 3), grid=True)
data.plot(y = 'days_exposition', kind = 'hist', bins = 100, grid = True, range = (0,100))

counts = data['days_exposition'][(data['days_exposition'] >= 0) & (data['days_exposition'] <= 100)]
values_to_replace = [0, 7, 30, 45, 60, 90]
median_values_by_week = data.groupby('week')['days_exposition'].median()

def replace_values(row):
    if row['days_exposition'] in values_to_replace:
        return median_values_by_week[row['week']]
    else:
        return row['days_exposition']
data['days_exposition'] = data.apply(replace_values, axis=1)
data.plot(y='days_exposition', kind='hist', bins=100, grid=True, range=(0, 600), figsize=(10, 6))
plt.title('Распределение времени размещения объявлений')
plt.xlabel('Время размещения (в днях)')
plt.ylabel('Количество объявлений')
plt.show()
data['days_exposition'].describe()
```

```python
mean_days_exposition = data['days_exposition'].mean()
median_days_exposition = data['days_exposition'].median()
print("Среднее значение длительности размещения объявления (в днях):", mean_days_exposition)
print("Медианное значение длительности размещения объявления (в днях):", median_days_exposition)
```

Создаем корреляционную матрицу
Сохраняем столбец last_price в отдельную таблицу с использованием copy()
Общая площадь квартиры имеет сильное влияние на цену недвижимости, 
что подтверждается высоким коэффициентом корреляции около 0.65.
Жилая площадь и площадь кухни также влияют на цену, однако их воздействие менее выражено,
с коэффициентами корреляции около 0.56 и 0.39 соответственно. 
Факторы, такие как тип этажа, день недели и год размещения объявления о продаже,
не имеют влияние на цену недвижимости

```python
floor_type_mapping = {'первый': 1, 'последний': 2, 'другой': 3}
data['floor_type_numeric'] = data['floor_type'].map(floor_type_mapping)
columns_to_analyze = ['last_price', 'total_area', 'living_area', 'kitchen_area', 'rooms', 'floor_type_numeric', 
                      'weekday_exposition', 'month_exposition', 'year_exposition']
corr_matrix = data[columns_to_analyze].corr()
last_price_corr = corr_matrix['last_price'].copy()
print(last_price_corr)
```

Найдем 10 населенных пунктов с наибольшим числом объявлений
Найдем среднюю цену за квадратный метр для этих населенных пунктов
Создадим сводную таблицу с количеством объявлений и средней ценой квадратного метра для этих населенных пунктов
Найдем населенный пункт с самой высокой средней ценой за квадратный метр
Найдем населенный пункт с самой низкой средней ценой за квадратный метр

```python
top_localities = data['locality_name'].value_counts().head(10).index
average_price_per_sqm = data[data['locality_name'].isin(top_localities)].groupby('locality_name')['price_per_sqm'].mean()
pivot_table = data[data['locality_name'].isin(top_localities)].pivot_table(index='locality_name', values='price_per_sqm', aggfunc=['count', 'mean'])
pivot_table.columns = ['count', 'mean']
pivot_table = pivot_table.sort_values('count', ascending=False)
locality_highest_price = average_price_per_sqm.idxmax()
locality_lowest_price = average_price_per_sqm.idxmin()
print(pivot_table)
print(f"Населенный пункт с самой высокой средней ценой за квадратный метр: {locality_highest_price}")
print(f"Населенный пункт с самой низкой средней ценой за квадратный метр: {locality_lowest_price}")
```

Посчитаем среднюю цену для каждого километра от центра
Построим график "Изменение средней цены квартир относительно расстояния до центра Санкт-Петербурга"

```python
spb_data = data[data['locality_name'] == 'Санкт-Петербург']
mean_price_per_km = spb_data.groupby('cityCenters_nearest_km')['last_price'].mean()
print(mean_price_per_km)
```

```python
mean_price_per_km = mean_price_per_km  
plt.figure(figsize=(10, 6))
plt.plot(mean_price_per_km.index, mean_price_per_km.values, marker='o', linestyle='-')
plt.title('Изменение средней цены квартир относительно расстояния до центра Санкт-Петербурга')
plt.xlabel('Расстояние до центра, км')
plt.ylabel('Средняя цена, млн руб.')
plt.grid(True)
plt.show()
```

Из полученных данных видно, что с увеличением расстояния от центра города Санкт-Петербурга до объекта недвижимости, средняя цена за квартиру уменьшается. Это может быть связано с тем, что квартиры, находящиеся ближе к центру города, обычно имеют более высокую стоимость из-за своего привлекательного расположения и близости к основным достопримечательностям и инфраструктуре.
Также можно заметить, что средняя цена за квартиру достигает пика при расстоянии в 0 км от центра города, а затем постепенно снижается. Это может быть связано с тем, что квартиры, находящиеся непосредственно в центре города, обычно имеют самую высокую стоимость из-за своего престижного расположения.
Мы видим на расстоянии от 4 до 6 км от центра города всплеск цены. Это может быть связанно с близостью к деловым и коммерческим районам, где много офисов и предприятий, а также из-за наличия престижных жилых комплексов.
На расстоянии 27 км от центра  так же есть  всплеск цены, это может быть связанно наличием крупных пригородных поселений или развитой инфраструктуры, которая делает этот район привлекательным для проживания.
Эти исключения указывают на то, что при анализе стоимости недвижимости необходимо учитывать не только расстояние от центра, но и другие факторы, которые могут влиять на цены в определенных районах.
